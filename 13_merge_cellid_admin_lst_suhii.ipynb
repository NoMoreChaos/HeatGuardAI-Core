```python
import pandas as pd

# cell_id, grid
df_grid = pd.read_csv("../data_processed/processed/grid/seoul_grid_100m.csv")

# cell_id, gu_name
df_gu_name = pd.read_csv("../data_processed/processed/admin_mapping/grid_to_gu.csv")

# cell_id, dong_name
df_dong_name = pd.read_csv("../data_processed/processed/admin_mapping/grid_to_dong.csv")

# cell_id, LST_C
df_lst = pd.read_csv("../data_processed/processed/lst/lst_2025_summer_100m_clean.csv")

# cell_id, SUHII
df_suhii = pd.read_csv("../data_processed/processed/suhii/suhii_2025_summer_100m.csv")

# cell_id, ta_chi
df_ta_chi = pd.read_csv("../data_processed/processed/ta_chi/tachi_with_grid.csv")

# cell_id, bus_pop
df_bus_pop = pd.read_csv("../data_processed/processed/bus_pop/grid_with_bus_population_2025.csv")

# cell_id, vulnerable
df_vulnerable = pd.read_csv("../data_processed/processed/vulnerable/grid_cellid_vulnerable_index.csv")

# cell_id, ndvi
df_ndvi = pd.read_csv("../data_processed/processed/ndvi/ndvi_2025_summer_100m.csv")
```

```python
df_grid = df_grid[["cell_id", "center_lat", "center_lon"]]
df_grid.info()
```

```python
df_lst.info()
```

```python
df_ndvi.info()
```

```python
df_suhii = df_suhii.rename(columns={"suhii": "SUHII"})
```

```python
def check_unique(df, key):
    return df[key].is_unique

print("grid:", check_unique(df_grid, "cell_id"))
print("gu_name:", check_unique(df_gu_name, "cell_id"))
print("dong_name:", check_unique(df_dong_name, "cell_id"))
print("lst:", check_unique(df_lst, "cell_id"))
print("suhii:", check_unique(df_suhii, "cell_id"))
print("ta_chi:", check_unique(df_ta_chi, "cell_id"))
print("bus_pop:", check_unique(df_bus_pop, "cell_id"))
print("vulnerable:", check_unique(df_vulnerable, "cell_id"))
print("ndvi:", check_unique(df_ndvi, "cell_id"))
```

```python
df = df_grid.copy()

df = df.merge(df_gu_name, on="cell_id", how="left")
df = df.merge(df_dong_name, on="cell_id", how="left")
df = df.merge(df_lst, on="cell_id", how="left")
df = df.merge(df_suhii, on="cell_id", how="left")
df = df.merge(df_ta_chi, on="cell_id", how="left")
df = df.merge(df_bus_pop, on="cell_id", how="left")
df = df.merge(df_vulnerable, on="cell_id", how="left")
df = df.merge(df_ndvi, on="cell_id", how="left")
```

```python
print(df.shape)
print(df.isna().mean().sort_values(ascending=False))
```

```python
df = df.rename(columns={
    "SIGUNGU_NM": "gu_name",
    "ADM_NM": "dong_name",
    "SIGUNGU_CD": "gu_cd",
    "ADM_CD": "dong_cd",
    "LST_C": "lst_c",
    "SUHII": "suhii",
    "유동인구_합계": "bus_pop",
    "vulnerable_index": "vulnerable",
    "NDVI": "ndvi"
})
```

```python
df_final = df[["cell_id", "center_lat", "center_lon", "gu_name", "dong_name", "lst_c", "suhii", "ta_chi", "bus_pop", "vulnerable", "ndvi"]]
```
